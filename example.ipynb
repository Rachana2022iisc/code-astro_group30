# Example notebook for the package

Given the RA/DEC (in ANY VALID format), the package will predict the class of the sources in the Field-of-View of the telescope.

The valid formats for RA/DEC are: hhmmss+-ddmmss and degree+-degree

As an example, we are going to use an object that is classified as a Quasar: SDSS J100554.68+312227.4
You can search the object at: https://simbad.u-strasbg.fr/simbad/sim-fbasic

`model_file` should contain the path to the trained `.joblib` file. You can either generate this file by running the machine-learning training code "trained_model.ipynb" once or download the post-trained `.joblib` file from Zenodo.

```python

from predict_sdss_stellar_class_1 import predict_sdss_coordinates

ra = "09h26m53.2549s"
dec = "+37d37m32.5018s"
search_radius = 1.0
result = predict_sdss_coordinates(
    ra=ra,
    dec=dec,
    radius_arcsec=search_radius,
    model_file="<path>/stellar_class_model.joblib"
)

result
```
